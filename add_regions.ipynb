```python
from geopy.geocoders import Nominatim
from tqdm import tqdm
import time

geolocator = Nominatim(user_agent="geo_region_finder")
def get_region(location_name):
    location = geolocator.geocode(location_name, exactly_one=True, addressdetails=True)

    if location and "address" in location.raw:
        address = location.raw["address"]
        # Get the most relevant administrative region
        if address.get("country_code") == "be":
            region = address.get("region", address.get("region", address.get("county", "Unknown region")))
            return region
        else:
            return "Unknown region"
    return "Unknown region"

tqdm.pandas()  # Enable progress bar
df["REGION"] = df["STATION"].progress_apply(lambda x: (time.sleep(1), get_region(x))[1])
```

```python
coordinate = pd.read_csv("coordinates.csv", sep=";")

coordinate = coordinate.melt(id_vars=['Geo Point'], value_vars=['Abréviation LST FR courte', 'Abréviation LST NL courte','Abréviation LST FR complète',
                                                                'Abréviation LST NL complète','Nom FR court','Nom NL court','Nom FR moyen','Nom NL moyen','Nom FR complet','Nom NL complet'], 
                      var_name='source_column', value_name='key_column')
coordinate.to_csv("test.csv", index=False)

df = pd.merge(df, coordinate, left_on='STATION', right_on='key_column', how='left')
df = df.drop(columns=['key_column', 'source_column'])
df = df.drop_duplicates(keep='first')


df[['Latitude', 'Longitude']] =df['Geo Point'].str.split(', ', expand=True)

df[['Latitude', 'Longitude']] = df[['Latitude', 'Longitude']].astype(float)

df = df.drop(columns=['Geo Point'])

df.to_csv("tempo.csv", index=False)
```
