```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefile
shapefile_path = "gadm41_BEL_2.shp"  # Update with the correct path if needed
belgium_provinces = gpd.read_file(shapefile_path)

belgium_provinces["centroid"] = belgium_provinces.geometry.centroid


fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
belgium_provinces.boundary.plot(ax=ax, color="black", linewidth=1)
belgium_provinces.centroid.plot(ax=ax, color="red", markersize=50)
plt.title("Centroids of Belgian provinces")
plt.show()

print(belgium_provinces[["NAME_2", "centroid"]])
```

```python
# Save the centroids to a new csv file

belgium_provinces[["NAME_2", "centroid"]].to_csv("belgium_provinces_centroids.csv", index=False)
print("Centroids saved to belgium_provinces_centroids.csv")
```
